```python
import cv2
```

```python
help(cv2)
```

```python
#how to read a image

image=cv2.imread('/content/cat.4065.jpg')
h,w=image.shape[:2]   # :2 --->
print(h,w)
```

```python
#how to display image

from google.colab.patches import cv2_imshow
cv2_imshow(image)
```

```python
#size increase

resize=cv2.resize(image,(600,600))
cv2_imshow(resize)
```

```python
roi=resize[100:200,200:700]
cv2_imshow(roi)
```

```python
roi=resize[100:220,240:400]
cv2_imshow(roi)
```

```python
w=600
h=600
center=(w//2,h//2)
matrix=cv2.getRotationMatrix2D(center,45,1.0)
rotated=cv2.warpAffine(resize,matrix,(w,h))
cv2_imshow(rotated)
```

```python
#copy image

copy_image=image.copy()
cv2_imshow(copy_image)
```

```python
#copy function

rectagle=cv2.rectangle(copy_image,(600,500),(200,200),(255,200,200))
cv2_imshow(rectagle)
```
